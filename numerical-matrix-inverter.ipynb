```python
import numpy as np
import time
```

```python
def invert_matrix(A): #See my notepad for derivation of algorithm
    A = np.array(A, dtype=np.float64) #This ensures the input, even if not already an np array, becomes an np array
    n_rows, n_cols = A.shape 
    n = n_rows
    if n_rows!= n_cols:
        raise ValueError("Matrix must be square to invert.")
    if n_rows == 1:
        if A[0,0] ==0:
            raise ValueError("Matrix is singular.")
        return np.array([[1/A[0,0]]])

    elif n_rows == 2:
        a, b = A[0,0] , A[0,1]
        c, d = A[1,0] , A[1,1]
        det = a*d - b*c
        if det == 0:
            raise ValueError("Matrix is singular.")
        inv = (1 / det) * np.array([[d,-b],
                                    [-c,a]])
        return inv

    else: n_rows > 2
    ddet = np.linalg.det(A) #Can eventually implement a formula using Levi-Civita def.
    if ddet == 0:
            raise ValueError("Matrix is singular.") 

    B = np.zeros((n,2*n), dtype=np.float64) #Creates a matrix with n rows and 2n columns rows from 0 to n-1 and columns from 0 to 2n-1 
    B[:, :n] = A # : means all rows here and :n means all columns 0 up to but not including n. Thus B[:,:n] refers to the left block of the augmented matrix (eg. the first n columns of it)
    B[:, n:] = np.eye(n) # n: means columns from n onward therefore this defines the right block and np.eye(n) creates the n x n identity matrix

    for i in range(n): #The for i in range(n) means 0,...,n-1 so up to but not including n which is precisely the number of rows 
        pivot = B[i,i]
        if pivot == 0:
            raise ValueError(f"Zero pivot encountered in row {i}.") #The f indicates an f-string. Python lets you insert variables directly inside curly braces
        B[i,:] = B[i,:] / pivot #B[i,;] represents the ith row rather than repeating this for each entry in the row by doing a loop over the j values in B[i,j] from j=0 to 2n-1

        for k in range(n): #This step encodes setting all other entries in the pivot column to 0 so only the 1 remains at b_ii in the left block
            if k != i:
                factor = B[k,i]
                B[k,:] = B[k,:] - factor*B[i,:]

    A_inv = B[:,n:] #Again refers to the right block as above, which now represents the inverse matrix
    return A_inv

    #At the time of writing this, there is a currently a flaw in the algorithm as if one of the pivots turns out to be 0, there is a chance this can be cured by a rearrangement of the matrix rows 
```
